```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
file_d = '/content/drive/MyDrive/Celerates MSIB/ALL Dataset/Mall_Customers.csv'
df = pd.read_csv(file_d)
df.head()
```

```python
df1 = df[["Age", "Spending Score (1-100)"]]
df1.head()
```

```python
model = DBSCAN(eps=5, min_samples=4)
model.fit(df1)
```

```python
df_new = df1.copy()
df_new['cluster'] = model.labels_
df_new.head()
```

```python
df_new.tail()
```

```python
plt.scatter(df_new['Age'], df_new['Spending Score (1-100)'], c=df_new['cluster'])

plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title('DBSCAN Clustering')
plt.show()
```

```python
anomalies = df_new[df_new['cluster'] == -1]
plt.scatter(df_new['Age'], df_new['Spending Score (1-100)'], c=df_new['cluster'])
plt.scatter(anomalies['Age'], anomalies['Spending Score (1-100)'], color='red')

plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title('DBSCAN Clustering with Anomalies')
plt.show()
```
